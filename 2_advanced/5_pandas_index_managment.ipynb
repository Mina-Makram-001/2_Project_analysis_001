```python
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df.index.name= 'job_id'
df.head()
```

```python
df_usa= df[df['job_country']=='United States']
df_usa.index.name= 'job_id before reset'
df_usa.reset_index(inplace=True)
df_usa.index.name= 'job_id after reset'
df_usa

```

```python
df_usa.set_index('job_id before reset', inplace=True)
df_usa.head()
```

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])

median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False, inplace=True)
median_pivot
```

```python
median_pivot.sort_index(ascending=True, inplace=True)
median_pivot
```
