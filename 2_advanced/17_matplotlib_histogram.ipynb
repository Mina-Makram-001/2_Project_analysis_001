## note
🔹 What is plt.gca()?

gca = Get Current Axes.

In Matplotlib, every plot lives inside a Figure → and a Figure can contain one or more Axes (subplots).

An Axes is basically the area where the graph is drawn (with x/y axis, ticks, labels, title, etc.).

So, when you call plt.gca(), Matplotlib gives you the current Axes object (the one being drawn on).

🔹 Workflow

Create a formatter → plt.FuncFormatter(lambda x, _: "custom string")

Apply it to the axis with set_major_formatter

🔹 1. lambda x, _: ...

This defines an anonymous function (a function without a name).

x → the tick value (e.g., 10000, 20000, …).

_ → the tick position (Matplotlib passes it, but here we ignore it, so _ is a Python convention meaning “unused variable”).

If we rewrote this with a normal function, it would be:

```
def my_formatter(x, pos):
    return f'${int(x/1000)}K'
```

````python
df_DA_US['salary_year_avg'].plot(kind='hist', bins=40, edgecolor='black')
plt.xlim(0, 250000)

ax = plt.gca() # get current axis
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.title('Distribution of United States Data Analyst Yearly Salaries')
plt.xlabel('Yearly Salary')
plt.ylabel('Number of Jobs')
````

## 🔹 Why can’t we just write `lambda x: ...`?

Because **Matplotlib’s `FuncFormatter` always calls your function with 2 arguments**:

```python
formatter_function(value, position)
```

* `value` → the tick value (`x`).
* `position` → the index of the tick (`pos`).

So if you wrote:

```python
plt.FuncFormatter(lambda x: f'${int(x/1000)}K')
```

Matplotlib would try to pass **two arguments** (`x` and `pos`) but your lambda only accepts **one**.
👉 That would cause a **TypeError**:

```
TypeError: <lambda>() takes 1 positional argument but 2 were given
```

---

## 🔹 Why does `lambda x, _` work?

* You declare **two parameters**: `x` and `_`.
* Matplotlib happily passes both (`value`, `position`).
* But since you don’t use the second one, you just call it `_` (a convention for "unused variable").

---

## 🔹 Alternative with `def`

```python
def my_formatter(x, pos):
    return f'${int(x/1000)}K'
```

Here `pos` is passed in, but you just don’t use it.

---

✅ **Summary**:

* `FuncFormatter` **requires a function with 2 arguments**.
* If you only write `lambda x: ...`, Matplotlib will crash because it tries to pass 2.
* Solution → either `lambda x, _: ...` or `def my_formatter(x, pos): ...`.
* `_` = "yes, I acknowledge the second argument, but I don’t care about it".
