### the methods to export your data to (excel, csv, sql, parquet, pickle,......etc) files

```python
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
months= df['job_posted_month'].unique()
```

```python
dict_monthes= {month: df[df['job_posted_month'] == month] for month in months}
dict_monthes['Jan'] # Example to access January data
```

```python
df_q1= pd.concat([dict_monthes['Jan'], dict_monthes['Feb'], dict_monthes['Mar']], ignore_index=True) #ignore_index=True to reset index
df_q1
```

### to_csv()

```python
df.to_csv('data_jobs.csv', index=False)
```

### to_excel()

```python
df_q1.to_excel('data_jobs.xlsx', index=False) 
```

### to_sql()

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///:memory:')  # Example using in-memory SQLite database ('sqlite:///:memory:') is the connection string for an in-memory SQLite database

df.to_sql('job_table', con=engine, if_exists='append', index=False)  # index=False to avoid writing the index to the table
```

### to_parquet()

```python
df_q1.to_parquet('data_jobs.parquet', index=False)  #index=False to avoid writing the index to the file
```

### to_pickle()

```python
df_q1.to_pickle('datajobs.pkl', index=False)  #index=False to avoid writing the index to the file
```
